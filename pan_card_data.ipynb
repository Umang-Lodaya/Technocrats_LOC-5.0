```python
import pandas as pd
from PIL import Image, ImageDraw, ImageFont
```

```python
df = pd.read_csv("/content/mock_data_500.csv")
```

```python
df
```

```python
pan_data = []
for i in range(df.shape[0]):
  pan_data.append(list(df.iloc[i, :]))
pan_data[:10]
```

```python
template = Image.open('pan2.png')
font = ImageFont.truetype("OpenSans-ExtraBold.ttf", size=20)
draw = ImageDraw.Draw(template)
draw.text((40, 120), pan_data[0][0], font=font, fill="black")
draw.text((40, 160), pan_data[0][1], font=font, fill="black")
draw.text((40, 200), pan_data[0][2], font=font, fill="black")
draw.text((40, 285), pan_data[0][3], font=font, fill="black")
```

```python
template.show()
```

```python
for i in range(len(pan_data)):
  template = Image.open('pan2.png')
  font = ImageFont.truetype("OpenSans-ExtraBold.ttf", size=20)
  draw = ImageDraw.Draw(template)
  draw.text((40, 120), pan_data[i][0], font=font, fill="black")
  draw.text((40, 160), pan_data[i][1], font=font, fill="black")
  draw.text((40, 200), pan_data[i][2], font=font, fill="black")
  draw.text((40, 285), pan_data[i][3], font=font, fill="black")
  template.convert('RGB').save(f"pan_data_imgs/{pan_data[i][0]}.jpg")
```
